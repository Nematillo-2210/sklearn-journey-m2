```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
coordinates, labels = make_blobs(n_samples=500, centers=4, cluster_std=0.8, random_state=42)
```

```python
clusters = [2, 3, 4, 5, 6, 7, 8]
results = []

for k in clusters:
    kmeans = KMeans(n_clusters=k) 
    kmeans_pred = kmeans.fit_predict(coordinates)
    silhouette = silhouette_score(coordinates, kmeans_pred)
    results.append(silhouette)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(clusters, results)
plt.xlabel('Clusters')
plt.ylabel('Results')
plt.title('comparing silhouette scores across different values of K')
plt.tight_layout()
plt.show()
```

```python
k_means2 = KMeans(n_clusters=4)
kmeans_pred2 = k_means2.fit_predict(coordinates)
samples= silhouette_samples(coordinates, kmeans_pred2)
```

```python
plt.figure(figsize=(20, 16))
y_start = 0 
for i in range(4): 
    filtered_samples = samples[kmeans_pred2 == i]
    sorted_samples = np.sort(filtered_samples)
    y_start_updated = y_start + len(sorted_samples)
    plt.barh(np.arange(y_start , y_start + len(sorted_samples)), sorted_samples)
    y_start = y_start_updated 
```
